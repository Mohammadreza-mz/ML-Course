# Q1
in this section we after importing needed libraries we read dataset form file into `df` dataframe.

we need years which have appeared in the columns so we save it into `years` list. we can do this by with writing every year in the list or extract it from dataframe column. for the second one we use `^\d+$` regex to match only column which thier names are a number.

also we need to multiply population by `1000` to make numbers correct. 
`df[years]` contains columns of dataframe which their label is a year so by multiplying that by 1000 we will acheive our goal.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HW1_Dataset.csv')
#years = ["1970", "1980", "1990", "2000", "2010", "2015", "2020", "2022"]
years = list(df.filter(regex="^\d+$").columns)
df[years] = 1000*df[years]
```

```python
df["Density"] = df['2022']/df['Area (km²)']
df
```

# Q2
`df[years].mean(axis=1)` returns average of columns which are in `years` for each row.
`axis=1` is written because mean function returns average for each column by default.

```python
df["avg"] = df[years].mean(axis=1)
df
```

# Q3
we create a list for column names which contains ratio to total population for each year. for example ratio for year `2022` will be `2022_ratio`. then we add this columns to dataframe. `df[years].sum()` returns total population for each year. for calculating ratio we simply divide each cell to total population and multiply it by `100` to keep it in percentage format.

```python
years_ratio = [x + "_ratio" for x in years]
df[years_ratio] = 100 * df[years] / df[years].sum()
df
```

# Q4
we add new column name `growth rate` to dataframe by dividing population in `2022` to population in `2020`.

```python
df["growth rate"] = df["2022"] / df["2020"]
df
```

# Q5
we keep current(2022) ration to total population in `2022_ratio`. so we need to sort rows by `2022_ratio` in descending mode.

```python
df = df.sort_values(by= "2022_ratio", ascending= False)
df
```

# Q6
we need to assign number to each row in the `Rank` column. we simply use their row number which is also unique.

```python
df["Rank"] = np.arange(df.shape[0])
df
```

# Q7
we write `binary_classification` function which will return desired string based on its input `x` which is `2022` population in some country.
by using `apply(binary_classification) on `df["2022"]` we will get the desired string for each row of dataframe and we save it into new column named `binary population class`

```python
def binary_classification(x):
    return "more than 300" if int(x) > 300000000 else "less than 300"
df["binary population class"] = df["2022"].apply(binary_classification)
df
```

# Q8
similar to Q7 we write another fuction that replaces the value `x` with `classification(x)` and we call that on `df["2022_ratio"]` and save the result into `population class` column.

```python
def classification(x):
    if float(x) < 2.0:
        return 0
    if float(x) >= 2.0 and float(x) < 5.0:
        return 1
    return 2

df["population class"] = df["2022_ratio"].apply(classification)
df
```

# Q9
`df.loc[df['2022'] > df['1970']]` finds rows where `df['2022'] > df['1970']`. we write `0` in the `growth status` columns of those rows.
we do the similar approach for rows where `df['2022'] <= df['1970']`.

```python
df.loc[df['2022'] > df['1970'], 'growth status'] = int(0)
df.loc[df['2022'] <= df['1970'], 'growth status'] = int(1)
df
```

# Q10
we find the row which it's `Name` is `China`. we only want population columns so we filter results with `filter(years)` and that only return columns which are in the `years` list.
then we sort `years` so that our X have ascending order.
our X will be `china_population.columns` and our Y will be their values. the shape of our result is `[1, len(years)]` but we want only one dimention so we reshape that into `len(years)` and then we plot the result.

```python
import matplotlib.pyplot as plt
china_population= df.loc[df['Name'] == "China"].filter(years)
china_population= china_population.reindex(sorted(china_population.columns), axis=1)
plt.plot(china_population.columns, china_population.to_numpy().reshape(len(years)))
```
